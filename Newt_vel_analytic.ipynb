# This code is used to calculate the flow rate and pressure drop in rectangular channel. This code originally written in MatLAB by Lichao Pan 05/17/2012 @ PENN. Updated/rewritten in Python by Alison Koser 12/07/12 @ PENN

```python
# Rectangular section: -a < y < a, -b < z < b
a = 250 # a = w/2 is half width, unit:um
b = 250 # b = h/2 is half height, unit:um

# Other Set Parameters
Q = 25.53333*10**7 # unit: um^3/s
mu = 0.624 # viscosity, unit: Pa s


############################################################
# Flow rate Q 
############################################################

K=0 # initial value set to zero

# Calculate the coefficient C_o
for i in range(1,9999,2):
    C_sum=tanh(i*pi*b/2/a)/(i**5)
    K=K+C_sum

C_o = 1-192*a/(pi**5)/b*K

dP_dx = Q/(4*b*a**3/3/mu)/C_o


############################################################
# Velocity profile u(y,z) @ z = 0
############################################################

u_y = linspace(-a,a,2*a+1)
y_y = linspace(0,2*a,2*a+1)
z = 0


############################################################
# Calculate Analytic Velocity Profile
############################################################

for y in range(-a,a):
    U_K = 0
    for j in range(1,999,2):
        u_sum =(-1)**((j-1)/2)*(1-cosh(j*pi*z/2/a)/cosh(j*pi*b/2/a))*(cos(j*pi*y/2/a)/j**3)
        U_K = U_K+u_sum   
    u_yz =16*a**2/mu/pi**3*dP_dx*U_K
    u_y[y+a+1]=u_yz/1000

    
# Plot Velocity Profile
plot(y_y,u_y, linewidth=3.5, color='black')
ylim(0,2.5)
xlim(0,500)
xlabel('channel position y  ($\mu$m)',fontsize=15)
ylabel('velocity (mm/s)',fontsize=15)


# Useful Values
v_max=max(u_y)
print 'Maximum Velocity: ', v_max, ' (mm/s)'
v_mean=mean(u_y)
print 'Average Velocity: ', v_mean, ' (mm/s)'
u_mean=Q/(4*a*b)*1000 #u_mean=Q/A
r_ratio=v_max/u_mean

```

```python
# Save output data: [channel position y, velocity(y)]
savetxt("Newt_Analytic.txt",
        array([y_y, u_y]).T)
```

```python
# Load experimental data
load_data=loadtxt('Newt_vel_analytic_v_steady.txt')
y_exp = load_data[:,0]#First Column
vel=load_data[:,1]#Second Column

# Plot analytic solution and experimental data together
plot(y_y,u_y, linewidth=5, color='black',label = 'Analyic Solution')
plot(y_exp,vel, marker='o', markersize=8,color='grey',
    linestyle='',label='Experimental Data')

ylim(0,2.4)
xlim(0,500)

xlabel('channel position y  ($\mu$m)',fontsize=15)
ylabel('velocity (mm/s)',fontsize=15)

legend(loc='lower center')
```
